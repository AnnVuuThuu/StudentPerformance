```python
x = df
y = df.iloc[:,8] #extract the 9th column
y
```

Data normalization

```python
from imblearn.over_sampling import RandomOverSampler
sm = RandomOverSampler()
x,y = sm.fit_resample(x,y)
```

train_test_split function = splits arrays or matices into random train and test subsets


parameters x = the feature matrix (input variables)
parameters y = the target vector (output variable)


test_size=0.3 = 30% dataset will be allocated to the test set and 70% will be used for training set


random_state=4 = guarantee that the split will be the same every time you run the code


X_train = The training set feature matrix

X_test = The test set feature matrix

y_train = The training set target vector

y_test = The test set target vector

```python
#splitting to test and train data
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=4)
```

KNN (K-Nearest Neighbors) assumes that similar data points exist close to each other in a given space. Therefore, our task is to find the k closest points to the data point we want to classify or predict

```python
#kneighbors
knn1=KNeighborsClassifier(algorithm='auto',n_neighbors=9,weights='distance')
knn1.fit(X_train,y_train)
y_pred1=knn1.predict(X_test)
print(classification_report(y_test,y_pred1))
```

Support Vector Machine (SVM) aims to find a hyperplane that separates data points. This hyperplane divides the space into different regions, where each region contains data points of a specific class

```python
#supportvector
sv=SVC(C=10, gamma =1, kernel= 'rbf')
sv.fit(X_train,y_train)
y_pred2=sv.predict(X_test)
y_pred2
print(classification_report(y_test,y_pred2))
```

Random Forest builds multiple decision trees, each trained on different subsets of the data. The final prediction is made based on a voting mechanism across all the decision trees

```python
#randomforest
rf=RandomForestClassifier(criterion= 'entropy', max_depth= None, min_samples_leaf= 1, min_samples_split= 4,n_estimators= 200)
rf.fit(X_train,y_train)
y_pred4=rf.predict(X_test)
y_pred4
print(classification_report(y_test,y_pred4))
```

```python
#decisiontree
rf=DecisionTreeClassifier(criterion= 'entropy', max_depth= None, min_samples_leaf= 1, min_samples_split= 4)
rf.fit(X_train,y_train)
y_pred4=rf.predict(X_test)
y_pred4
print(classification_report(y_test,y_pred4))
```

XGBoost (Extreme Gradient Boosting) is a machine learning technique used to minimize errors during predictive data analysis. It offers fast learning speed and high performance

```python
#xgboost
xgb=XGBClassifier()
xgb.fit(X_train,y_train)
y_pred7=xgb.predict(X_test)
y_pred7
print(classification_report(y_test,y_pred7))
```
